# outlier handling in price column

```python

df['price'].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.histplot(df['price'], kde = True)
plt.subplot(1,2,2)
sns.boxplot(df['price'])
plt.show()


```

```python
lower_cap = df['price'].quantile(0.05)
upper_cap = df['price'].quantile(0.95)
```

```python
lower_cap
```

```python
upper_cap
```

```python
df['price_capped'] = np.where(df['price'] < lower_cap, lower_cap,
         np.where(df['price'] > upper_cap, upper_cap, df['price']))

df['price'] = df['price_capped']
df.drop('price_capped', axis=1, inplace = True)
```

```python
df['price'].describe()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.histplot(df['price'], kde = True)
plt.subplot(1,2,2)
sns.boxplot(df['price'])
plt.show()
```

```python
df.head()
```

```python
df.columns
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df['price'] = df['price'].clip(lower = lower_bound, upper = upper_bound)
```

```python
df['price'].describe()
```

```python
sns.boxplot(x= 'price', data = df)
```

## outlier handling in battery_capacity column

```python
df['battery_capacity'].describe()
```

```python
Q1 = df['battery_capacity'].quantile(0.25)
Q3 = df['battery_capacity'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

outliers = df[(df['battery_capacity'] < lower_bound) | (df['battery_capacity'] > upper_bound)]
print("Outliers:\n", outliers[['model', 'battery_capacity']])
```

```python
df['battery_capacity'] = df['battery_capacity'].clip(lower=lower_bound, upper=upper_bound)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['battery_capacity'])
plt.title("Battery Capacity after Outlier Handling")
plt.show()

print(df['battery_capacity'].describe())
```

 Q1. Which smartphone brands are regarded as the most popular?

```python
df.head()
```

```python

import plotly.graph_objects as go
import plotly.express as px
```

```python
mobile_brands = df.groupby('brand_name').agg({'brand_name':'count','avg_rating' : 'mean'}).rename(columns={'brand_name': 'brand_count'}).sort_values(by = 'brand_count', ascending =False)
mobile_brands

```

```python
plt.figure(figsize=(17,6))
sns.barplot(x=mobile_brands.index, y=mobile_brands['brand_count'], palette='viridis')
plt.title('Top 10 Most Popular Smartphone Brands (by Count of Models)')
plt.xlabel('Brand Name')
plt.ylabel('Number of Models')
plt.xticks(rotation=45)
plt.show()
```

**So xiaomi and samsung are most popular brand names.**

Q2. Which brands among the top 10 receive high ratings?

```python
top10_brands = mobile_brands.head(10)
top10_brands
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x=top10_brands.index, y=top10_brands['avg_rating'], palette='viridis')
plt.title('Top 10 Most Popular Smartphone Brands (by Average Rating)')
plt.xlabel('Brand Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

- Oneplus and Motorola receive high ratings among the top 10 brands.

Q3. Which brand offer expensive smartphones, and is there any correlation between the ratings and prices of smartphones?

```python
# Average price per brand
brand_price = df.groupby('brand_name').agg({'price':'mean','avg_rating':'mean'}).sort_values(by='price', ascending=False)
brand_price  # Top 10 expensive brands
```

```python
plt.figure(figsize=(17,6))
sns.scatterplot(x=brand_price.index, y=brand_price['price'], palette='viridis')
plt.title('Average Price of Smartphones by Brand')
plt.xlabel('Brand Name')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='price', y='avg_rating', data=df)
plt.xlabel('Price')
plt.ylabel('Average Rating')
plt.title('Correlation between Price and Ratings of Smartphones')
plt.show()
```

Q4. Among the various smartphone brands available in the market, which brand is known to offer the highest number of 5G smartphones?

```python
df['5G_or_not'].value_counts()
```

```python
# Filter only 5G smartphones
df_5G = df[df['5G_or_not'] == 1]  # 1 means 5G

# Count 5G models per brand
brand_5G_count = df_5G['brand_name'].value_counts()
top_10_5G = brand_5G_count.head(10)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.pie(top_10_5G.values,
        labels=top_10_5G.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.tab20.colors)
plt.title('Top 10 Brands Offering Most 5G Smartphones')
plt.show()
```

- So, xiaomi offer the highest number of 5G smartphones
